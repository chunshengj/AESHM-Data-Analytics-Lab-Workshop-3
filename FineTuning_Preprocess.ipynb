## Convert csv into jsonl

```python
def convert_jsonl(df, output_file):
    system_message = {
        "role": "system",
        "content": "Your task is to classify consumer hotel review into one of the following ratings: [\"1\", \"2\", \"3\", \"4\", \"5\"], where \"1\" indicates very dissatisfied and \"5\" indicates very satisfied."
        }

    with open(output_file, "w", encoding="utf-8") as f:
        for _, row in df.iterrows():
            entry = {
                "messages": [
                    system_message,
                    {"role": "user", "content": str(row["review_text"])},
                    {"role": "assistant", "content": str(row["value"])}  # ensure label is string
                ]
            }
            f.write(json.dumps(entry, ensure_ascii=False) + "\n")  # keep emojis
```

```python
convert_jsonl(df, "fine_tune_VALUE_data.jsonl")
```

# Read jsonl file

```python
file_path = "fine_tune_VALUE_data.jsonl"

examples = []
with open(file_path, "r", encoding="utf-8") as f:
    for line in f:  
        examples.append(json.loads(line))  

print(f"Total examples: {len(examples)}")
examples[0:5]
```
